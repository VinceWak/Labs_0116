#### Read the `employees` dataset
Save your data in a variable called `employees`.

```python
employees = pd.read_csv("Employee.csv")
employees
```

#### What's the mean `Salary` by `Department`?
**Tip**: There are 2 ways to do it: with or without pivot tables. Do it both ways. 

```python
employees.groupby("Department").mean()
```

```python
employees.pivot_table(index=["Department"])
```

#### Now, group the data by more than one column. Use Pandas pivot table function to select the mean `Salary` by `Department` and `Title`.

```python
employees.pivot_table(index=["Department", "Title"], values='Salary')
```

#### Display the number of `Years` of work experience and the `Salary` for each `Title`. Add `Years` as a column in your pivot table.
**Hint**: Use Pandas documentation to explore the [pivot table](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.pivot_table.html) function.

```python
employees.pivot_table(index=('Title', 'Years'), values='Salary')
```

#### Find the mean `Salary` and number of employees in each `Department` and for each job `Title`.

```python
employees.pivot_table(index=('Department', 'Title'), values=['Salary', 'Name'], aggfunc={'Salary':np.mean, 'Name':np.count_nonzero})
```

#### For each `Department` and each job `Title`, find the `Salary` median, minimum, maximum and standard deviation. Comment your results.

```python
employees.pivot_table(index=('Department', 'Title'), values=['Salary'], aggfunc=[np.median, np.min, np.max, np.std])
```

```python
# The resulting pivot table will show the median, minimum, maximum, and standard deviation of salary for each department and title
```

#### Based on your comments, fill the missing values with an appropriate value.

```python
employees.pivot_table(index=('Department', 'Title'), values=['Salary'], aggfunc=[np.median, np.min, np.max, np.std], fill_value=0)
```

#### The stake holders want to know, for each `Department`, the number of employees and how much money is spent on salaries. Could you provide that information? 

```python
employees.pivot_table(index=('Department'), values=['Name', 'Salary'], aggfunc={'Name':np.count_nonzero, 'Salary':np.sum})
```

#### For each `Department` and `Title`, find the the total years of work experience and the mean `Salary`.

```python
employees.pivot_table(index=('Department', 'Title'), values=['Years', 'Salary'], aggfunc={'Years':np.sum, 'Salary':np.mean})
```
